# Data Processing

```python
pd.set_option('display.max_columns',None)
def data_quality_check(df, df_name="DataFrame"):
    """
    Menampilkan metrik kualitas data dari sebuah DataFrame.

    Parameters:
    df (pd.DataFrame): DataFrame yang akan diperiksa kualitasnya.
    df_name (str): Nama DataFrame untuk label output cetak. Default 'DataFrame'.

    Returns:
    None (Hanya mencetak statistik berupa bentuk data, baris duplikat, 
          missing values, dan tipe data ke konsol).
    """

    print("=" * 50)
    print(f"CHECKING : {df_name}")
    print("=" * 50)

    print(f"Shape : {df.shape}")

    # Duplicate
    dup_count = df.duplicated().sum()
    print(f"\nDuplicate Rows : {dup_count}")

    # Missing Value
    missing = df.isna().sum()

    print("\nMissing Values:")
    print(missing[missing > 0])

    # Data Type
    print("\nData Types:")
    print(df.dtypes)

    print("\n")
```

## Product

```python
# read file csv
product_table = pd.read_csv('products.csv')

# check kualitas data
data_quality_check(product_table,'Product')
product_table.head()
```

Temuan:
- tidak memiliki duplicated
- memiliki missing value column rating_avg dengan jumlah 548
- inconsissten dtype pada column date_added
- belum lowwercase

```python
# membandingkan product_table['rating_avg'] 
display(product_table.loc[product_table['rating_avg'].isna()].head())
display(product_table.loc[product_table['rating_avg']>0].head())
```

Temuan:

dari perbandingan kedua dataframe tersebut terjadinya kondisi missing value pada ranting_avg terjadi dikarnakan tidak adanya review terhadap product tersebut, kemungkinan:
- product baru
- atau product kurang diminati 

```python
# mengisi missing value dengan angka 0
product_table['rating_avg'] = product_table['rating_avg'].fillna(0)

# Mengubah dtype colomn date_added
product_table['date_added'] = pd.to_datetime(product_table['date_added'],yearfirst=True)

# Menormalisasi text menjadi lowercase
product_objc = product_table.select_dtypes(include='object').columns
for col in product_objc:
    product_table[col] = (
        product_table[col]
        .str.strip()
        .str.lower()
        .str.replace('_',' ')
    )
```

```python
# memfilter ascending berdasarkan date_add
product_table.sort_values(by='date_added',inplace=True)

# membuat column baru 
product_table['newproduct_id']=product_table.index+1

# mendelete index
product_table.reset_index(drop=True,inplace=True)
product_table.head()
```

```python
product_table.info()
```

```python
# Mengcoppy dataframe
df_products = product_table.copy()

# menimpa product_id baru
df_products['product_id'] = df_products['newproduct_id']

# mendelete newproduct_id
df_products.drop(columns=['newproduct_id'],inplace=True)
df_products.sample(n=10,random_state=11)
```

```python
df_products.info()
```

```python
# export dataframe product -> csv
df_products.to_csv('dataframe_product.csv',index=False)
```

## User

```python
user_table = pd.read_csv('users.csv')
data_quality_check(user_table,"User")
```

```python
user_table['signup_date'] = pd.to_datetime(user_table['signup_date'],yearfirst=True)
user_objc = user_table.select_dtypes(include='object').columns

for col in user_objc:
    user_table[col] = (
        user_table[col]
        .str.strip()
        .str.lower()
        .str.replace('_',' ')
    )
user_table.sort_values(by='signup_date',inplace=True)
user_table['newuser_id'] = user_table.index+1
user_table.reset_index(drop=True,inplace=True)
user_table.sample(n=10, random_state=11)
```

```python
user_table.info()
```

```python
df_users = user_table.copy()
df_users['user_id'] = df_users['newuser_id']
df_users.drop(columns=['newuser_id'], inplace= True)
df_users.sample(random_state=11,n=500)
```

```python
df_users.info()
```

```python
df_users.to_csv('dataframe_user.csv',index=False)
```

## Session

```python
session_table = pd.read_csv('sessions.csv')
data_quality_check(session_table,"Session")
```

```python
session_table['start_time'] = pd.to_datetime(session_table['start_time'],yearfirst=True)
session_objc = session_table.select_dtypes(include='object').columns

for col in session_objc:
    session_table[col] = (
        session_table[col]
        .str.strip()
        .str.lower()
        .str.replace('_',' ')
    )
session_table.sort_values(by='start_time',inplace=True)
session_table['newsession_id'] = session_table.index+1
session_table.reset_index(drop=True,inplace=True)
session_table = session_table.merge(
    user_table,
    how='inner',
    on='user_id'
)
```

```python
session_table.head()
```

```python
session_table.info()
```

```python
df_sessions = session_table.copy()
df_sessions[['session_id','user_id']] = df_sessions[['newsession_id','newuser_id']]
df_sessions.drop(columns=['newsession_id','newuser_id'],inplace=True)
df_sessions.sample(n=10,random_state=11) 
```

```python
df_sessions.info()
```

```python
df_sessions.to_csv('dataframe_session.csv',index=False)
```
